```python

def fact(n):
    f = 1
    if n == 0 :
        print('1')
    else:
        for i in range(1, n+1):
             f *= i
        print(f)

fact(3)
```

```python
def fact_rec(n):
    if n == 1:
        return 1
    else:
        return n * fact_rec(n-1)

print(fact_rec(3))
```

```python
def mul(x,y):
    if y == 0:
        return 0
    elif y == 1:
        return x
    else:
        return x + mul(x,y-1)
    
print(mul(2,3))
```

```python
def pow_rec(x,y):
    if y == 0:
        return 1
    elif x == 0:
        return 0
    elif y == 1:
        return x
    else:
        return x * pow_rec(x,y-1)
    
print(pow_rec(3,2)) 
```

```python
def fibo(n):
    r = []
    a = 0
    b = 1
    while a < n:
        r.append(a)
        a, b = b, a+b
    return r

print(fibo(10))
```

```python
def fibo(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

print(fibo(4))   # 
```

```python

def f(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return lst[0] + f(lst[1:])    

a = [2, 4, 5, 6, 7]
print(f(a))
```

```python
def sum_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_digits(int(n/10)) 

print(sum_digits(345))
```

```python
def sum_series(n):
    if n < 1 :
        return 0
    else:
        return n + sum_series(n-2)

print(sum_series(10)) 
```

```python
def f(n,base):
     s = '0123456789ABCDEF'
     if n < base:
         return s[n]
     else:
         return f(n//base , base) + s[ n % base]

print(f(10,16))       
print(f(25,16))
```

```python
def binary_search(lst, x, start=0, end=None):
    if end is None:
        end = len(lst) - 1
    if start > end:
        return False
    mid = (start + end) // 2
    if x == lst[mid]:
        return mid
    if x < lst[mid]:
        return binary_search(lst, x, start, mid - 1)
    return binary_search(lst, x, mid + 1, end)

a = [2, 4, 7, 12, 19, 25, 38]
print(binary_search(a, 19 ))    
print(binary_search(a, 4 ))    
print(binary_search(a, 20)) 
```
