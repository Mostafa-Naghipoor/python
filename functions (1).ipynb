```python
count_dict = {
         'L' : 0 ,
         'U' :0
        }

def func_count(s):
    for ch in s:
        if ch.islower():
            count_dict['L'] += 1
        else:
            count_dict['U'] += 1

s = 'MosTAfa'            
func_count(s)            
print(count_dict['L'])  
print(count_dict['U'])
```

```python
def count_char(s):
    d = {}
    for i in s:
        if i in d.keys():
            d[i] +=1
        else:
            d[i] = 1
    return d

print(count_char('abbcfab'))
```

```python
def count_word(s):
    d = {}
    words = s.split()
    for i in words:
        if i in d:
            d[i] += 1
        else:            
            d[i] = 1
    return d
            
s = 'python created by rossum'
print(count_word(s))
```

```python
grade_student = [ 
                  {'id':1 , 'M':60 , 'F':40} , 
                  {'id':2 , 'M':80 , 'F':70}
                ]

def ave_grade(lst):
    for d in lst:
        n1 = d.pop('M')
        n2 = d.pop('F')
        d['Ave'] = (n1 + n2) / 2
    return lst
```

```python
print(ave_grade(grade_student))
```

```python
def remove_index(s,start,end):
    if len(s) > end:
        s = s[0: start] + s[end+1 ::]
    return s    

s= 'python'
print(remove_index(s,1,3))
```

```python

def remove_oddindex(s):
    r = ''
    for i in range(len(s)):
        if i % 2 == 0:
            r += s[i]
    return r            

print(remove_oddindex('python'))    
print(remove_oddindex('abcdefgh'))
```

```python
def prime(n):
    my_set = set()
    lst = []
    for i in range(2, n+1):
        if i in my_set:
            continue
        for j in range(i*2, n+1, i):
            my_set.add(j)     
        lst.append(i)         
    return lst        
n = 10
print(prime(n)) 
```

```python
def magic(m):
    n = len(m[0])
    lst = []
    
    lst.extend([sum(i) for i in m])
    
    for col in range(n):
        lst.append(sum(row[col] for row in m))
    
    r = 0
    for i in range(0,n):
        r += m[i][i]
    
    lst.append(r)
    
    r2 = 0
    for i in range(0,n):
        r2 += m[i][n-1-i]
    
    lst.append(r2)
    if len(set(lst)) > 1:
        return False
    return True
     
my_matrix = [
     [2, 7, 6],
     [9, 5, 1],
     [4, 3, 8]
    ]

print(magic(my_matrix)) 
```
